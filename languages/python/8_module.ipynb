## Introduction to NumPy

NumPy is a Python library used for working with arrays. It also has functions for working in domain of linear algebra, Fourier transform, and matrices.

```python
# If running this notebook in an environment where these libraries are not installed, uncomment the lines below to install them
# !pip install numpy
# !pip install pandas
# !pip install matplotlib
```

```python
import numpy as np

# Create a numpy array
arr = np.array([1, 2, 3, 4, 5])
print("Numpy array:", arr)

# Do some array operations
print("Array + 2:", arr + 2)
print("Array * 2:", arr * 2)
```

## Introduction to Pandas

Pandas is a library used for data manipulation and analysis. It is used to extract data and store it in the desired format.

```python
import pandas as pd

# Create a pandas dataframe
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

print("Pandas DataFrame:")
print(df)
```

## Introduction to Matplotlib

Matplotlib is a plotting library. Matplotlib is a low-level library with a Matlab like interface which offers lots of freedom at the cost of having to write more code.


```python
import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('some numbers')
plt.show()
```
